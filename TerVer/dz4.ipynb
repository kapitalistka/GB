<b>Решение:</b><br>
для равномерного распределения дисперсия вычисляется по формуле:<br>
$$D(X) = \dfrac{(b - a)^2}{12}$$

известны D=0.2 и a=0.5

подставим:

$$0.2=\dfrac{(b - 0.5)^2}{12}$$

$$2.4={(b - 0.5)^2}, b^2-b-2.15=0$$

$$b1=-1.04, b2=2.04$$


Тк мы ищем правую границу, то нас интересует число, больше а, следовательно b=2.04

$$M(X)=\dfrac{a+b}{2} = 1.27$$
