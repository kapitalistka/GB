1. Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
def add_vectors(a: list, b: list):
    if len(a)!=len(b): 
        raise Exception("Different dimensions")        
    else:
        c= [ai+bi for ai,bi in zip(a,b)]
        return c
        
        
           
    
```

```python
add_vectors([10,10,10] , [0,0,-10])
```

```python
add_vectors([3,5,2,8],[3,8,2,8])
```

3. Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
x=[]
y1=[]
y2=[]
p=5
for i in range(0,100):
    x_=i/1000
    x.append(x_)
    y1.append(math.sqrt(2*p*x_))
    y2.append(-math.sqrt(2*p*x_))

    
plt.plot(x,y1)
plt.plot(x,y2)

plt.xlabel("x")
plt.ylabel("y")
```

```python
def pol_to_dec(r, a):
    x = r * np.cos(a)
    y = r * np.sin(a)
    return(x, y)
```

4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
x = np.linspace(-5,5,100)
y1=x**2
y2=(np.exp(x)+x-1)/x
plt.plot(x, y1)
plt.plot(x,y2)
plt.grid(True)
```

```python
def equation(p):
    x, y =p
    return (y-x**2+1,y-(np.exp(x)+x-1)/x)


#из графика видно, что система имеет 3 корня:
x1,y1= fsolve(equation, (-2,0))
print(x1,y1)
x2,y2= fsolve(equation, (1,1))
print(x2,y2)
x3,y3= fsolve(equation, (4,15))
print(x3,y3)
```
